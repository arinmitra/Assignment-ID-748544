```python
# Exercise 1: Automating Sales Data Analysis with Copilot
# Task 3: Exploratory Analysis – Key Questions
# 
"""
o Which branch has the highest total sales?
○ What are the top-selling product categories?
○ Do members spend more on average than normal customers? 
Compare average purchase totals between Member vs Normal customers. You can 
do this by grouping by Customer type and calculating the mean of the Total column.
○ This will show the average sale amount for Members and for Normal customers. 
Note which is higher. 
(This addresses whether the membership program is associated with higher 
spending per transaction.)
○ How do customer ratings differ by branch?
○ If you want to explore the Rating data: group by Branch 
and find the average customer satisfaction rating for each branch. 
(Use a similar approach with df.groupby('Branch')['Rating'].mean().)
○ This can tell if one store branch has consistently higher or lower 
customer feedback. If one branch has lower ratings, it might need 
attention to customer service.
""" 
# 
import pandas as pd
# Load the sales data
file_path = 'cleaned_supermarket_sales.csv'
df = pd.read_csv(file_path)

# 1. Which branch has the highest total sales?
highest_sales_branch = df.groupby('Branch')['Total'].sum().idxmax()
highest_sales_value = df.groupby('Branch')['Total'].sum().max()
print(f"Branch with highest total sales: {highest_sales_branch} (${highest_sales_value:.2f})")

# 2. What are the top-selling product categories?
top_selling_categories = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
print("\nTop-selling product categories:")
print(top_selling_categories)

# 3. Do members spend more on average than normal customers?
average_spending = df.groupby('Customer type')['Total'].mean()
print("\nAverage spending by customer type:")
print(average_spending)
if average_spending['Member'] > average_spending['Normal']:
    print("Members spend more on average than Normal customers.")
else:
    print("Normal customers spend more on average than Members.")

# 4. How do customer ratings differ by branch?
average_ratings_by_branch = df.groupby('Branch')['Rating'].mean()
print("\nAverage customer ratings by branch:")
print(average_ratings_by_branch)

# 5. If one branch has lower ratings, it might need attention to customer service.
lowest_rating_branch = average_ratings_by_branch.idxmin()
lowest_rating_value = average_ratings_by_branch.min()
print(f"\nBranch with lowest average rating: {lowest_rating_branch} ({lowest_rating_value:.2f})")

# This branch may need attention to customer service.
if lowest_rating_value < 3.0:
    print(f"Attention needed for {lowest_rating_branch} branch due to low customer ratings.")
else:
    print(f"{lowest_rating_branch} branch has acceptable customer ratings.")

# End of the script
```
