```python
# Exercise 1: Automating Sales Data Analysis with Copilot
# Task 4: Visualization – Creating Charts
# 
"""
This script demonstrates how to create a bar chart using matplotlib or seaborn 
to visualize sales data.

Next, create some visualizations to include in your report/dashboard. 
Use Copilot to help generate plotting code. Remember to label your axes 
and add titles for clarity. 
(You can use matplotlib or seaborn for plots. 
Seaborn often produces prettier default styles 
– you can import seaborn as sns if you like.)

a. Bar chart of total sales by branch:
○ Using the result from which branch has the highest total sales [3(a)], 
plot a bar chart to compare the total sales of Branch A, B, and C.
○ Ensure the chart displays three bars for branches A, B, C with their 
respective total sales. The highest bar should correspond to the 
branch you identified earlier.
○ If Copilot doesn’t add labels or titles, make sure to add them 
yourself for a professional presentation. You can also customize 
colors as shown or leave them default.
"""

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data for total sales by branch
file_path = 'cleaned_supermarket_sales.csv' # Adjust the file path as needed
data = pd.read_csv(file_path)
# Calculate total sales by branch 
total_sales_by_branch = data.groupby('Branch')['Total'].sum().reset_index()
# Sort the data by total sales in descending order
total_sales_by_branch = total_sales_by_branch.sort_values(by='Total', ascending=False)
# Create a bar chart using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6)) 
sns.barplot(x='Branch', y='Total', data=total_sales_by_branch, palette='viridis')
# Add labels and title 
plt.xlabel('Branch')
plt.ylabel('Total Sales') 
plt.title('Total Sales by Branch')
# Show the plot
plt.tight_layout()
plt.show()

# Save the plot as an image file
plt.savefig('total_sales_by_branch.png', dpi=300) 


```
