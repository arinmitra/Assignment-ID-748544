```python
# Exercise 1: Automating Sales Data Analysis with Copilot
# Task 4b: Bar chart of total sales by product line:
# 
""" This script generates a bar chart of total sales by product line 
using matplotlib.
It reads sales data from a CSV file, processes it to calculate total sales
by product line, and then creates a bar chart to visualize the results.

 Visualize the sales for each product category. There are multiple product 
lines (e.g., Food & Beverages, Fashion accessories, etc.).
 Because there are several categories, you may want to make the figure 
larger and rotate x-axis labels for readability.
 This should produce a chart where each bar represents a product categorys 
total sales. From this, identify which product line is highest 
(it should match your earlier fi nding).
"""

import pandas as pd
import matplotlib.pyplot as plt

# Load the sales data from a CSV file
file_path = 'cleaned_supermarket_sales.csv' 
sales_data = pd.read_csv(file_path)

# Group the data by product line and calculate total sales for each line
total_sales_by_product_line = sales_data.groupby('Product line')['Total'].sum().reset_index()
# Sort the data by total sales in descending order
total_sales_by_product_line = total_sales_by_product_line.sort_values(by='Total', ascending=False)
# Create a bar chart to visualize total sales by product line
plt.figure(figsize=(12, 6))  # Set the figure size for better readability
plt.bar(total_sales_by_product_line['Product line'], 
        total_sales_by_product_line['Total'], color='skyblue')
plt.title('Total Sales by Product Line', fontsize=16)
plt.xlabel('Product Line', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()  # Display the bar chart
# Save the bar chart as an image file
plt.savefig('total_sales_by_product_line.png', bbox_inches='tight')
# Print the product line with the highest total sales
highest_sales_product_line = total_sales_by_product_line.iloc[0]
print(f"The product line with the highest total sales is: {highest_sales_product_line['Product line']} "
      f"with a total of {highest_sales_product_line['Total']:.2f}")

```
