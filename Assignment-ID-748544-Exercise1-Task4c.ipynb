```python
# Exercise 1: Automating Sales Data Analysis with Copilot
# Task 4c: Comparison of member vs non-member spending
# 
"""
This script compares the average spending of members and non-members
in a sales dataset. It calculates the average spending for each group
and visualizes the results using a bar chart.

○ If you did step 3(c), you can create a simple bar chart for average spending 
of Members vs Normal customers. This is a quick two-bar comparison.
○ The chart will help visualize the diff erence in average sale amount, 
complementing the numeric values you computed.

"""
import pandas as pd
import matplotlib.pyplot as plt

# Load the sales data from a CSV file
file_path = 'cleaned_supermarket_sales.csv' 
sales_data = pd.read_csv(file_path)

# Calculate average spending for members and non-members
average_spending = sales_data.groupby('Customer type')['Total'].mean()
# Prepare data for plotting
average_spending_data = average_spending.reset_index()
# Create a bar chart to compare average spending
plt.figure(figsize=(6, 4))
plt.bar(average_spending_data['Customer type'], average_spending_data['Total'], width=0.3, color=['blue', 'orange'])
plt.title('Average Spending: Members vs Non-Members')
plt.xlabel('Customer Type')
plt.ylabel('Average Spending ($)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
# Show the plot
plt.show()
# Save the plot as an image file
plt.savefig('average_spending_comparison.png')
# Print the average spending values
print("Average Spending Comparison:")
print(average_spending_data.to_string(index=False))

# Save the average spending data to a CSV file
average_spending_data.to_csv('average_spending_comparison.csv', index=False)
# Print a message indicating that the analysis is complete
print("Average spending comparison analysis complete. Results saved to 'average_spending_comparison.csv' and plot saved as 'average_spending_comparison.png'.") 
# End of script 
```
