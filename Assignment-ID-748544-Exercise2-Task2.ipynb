```python
#########################################################
# This file is part of the assignment for the course    #
# "AI Driven Data Analytics Course" by Skillarbitrage   #
# Assignment ID: 748544                                 #
# Submissions of weekly Exercises for                   #
# Exercises on AI-Powered Automation for Data Science   # 
# Using GitHub Copilot in VS Code                       #
# Week Duration: 15/06/2025 to 22/06/2025               #
#                                                       #
# Student Name : Arindam Mitra                          #
# Registration Code: STU20250430/14045                  #
# Email: arinmitra@gmail.com                            #
#                                                       #
# Tutor - Mohit Sharma (mohit.sharma@skillarbitra.ge),  #
# Host - Ritu Verma (ritu.varma@skillarbitra.ge)        #
#########################################################

# Exercise 2: Creating an Automated Climate Data Dashboard with Copilot
# Task 2: Date Parsing and Sorting
""" 
This script is designed to parse and sort climate data from a CSV file.

○ Convert the date column from string to datetime objects. 
This is crucial for time-series analysis.
○ Ensure the DataFrame is sorted by date in ascending order. The data might 
already be in order, but it’s good practice to sort after conversion.
○ Set the date column as the DataFrame index (optional, but useful for time-series plotting). 
This will allow easier resampling if needed and nicer plotting on the x-axis.
"""

import pandas as pd
from datetime import datetime

# Load the climate data from a CSV file
file_path = 'DailyDelhiClimateTest.csv'
climate_data = pd.read_csv(file_path)

# Display the first few rows of the dataset before conversion
print("First few rows of the climate data before date conversion:")
print(climate_data.head())
# Display the data types of each column before conversion
print("\nData types of each column before date conversion:")
print(climate_data.dtypes)

# Convert the 'date' column from string to datetime objects
climate_data['date'] = pd.to_datetime(climate_data['date'], format='%Y-%m-%d')
# Display the first few rows of the dataset after conversion
print("\nFirst few rows of the climate data after date conversion:")
print(climate_data.head())
# Display the data types of each column after conversion
print("\nData types of each column after date conversion:")
print(climate_data.dtypes)

# Ensure the DataFrame is sorted by date in ascending order
climate_data.sort_values(by='date', inplace=True)
# Display the first few rows of the sorted dataset
print("\nFirst few rows of the sorted climate data:")
print(climate_data.head())

# Set the date column as the DataFrame index
climate_data.set_index('date', inplace=True)
# Display the first few rows of the DataFrame with date as index
print("\nFirst few rows of the climate data with date as index:")
print(climate_data.head())
# Display the index of the DataFrame
print("\nIndex of the DataFrame:")
print(climate_data.index)
# Save the processed DataFrame to a new CSV file
output_file_path = 'Processed_DailyDelhiClimate.csv'
climate_data.to_csv(output_file_path)
print(f"\nProcessed climate data saved to {output_file_path}")
# Display the final DataFrame
print("\nFinal processed climate data:")
print(climate_data)
# Display the shape of the final DataFrame
print("\nShape of the final processed climate data:")
print(climate_data.shape)
# Display the first few rows of the final DataFrame
print("\nFirst few rows of the final processed climate data:")
print(climate_data.head())
# Display the last few rows of the final DataFrame
print("\nLast few rows of the final processed climate data:")
print(climate_data.tail())



```
