```python
#########################################################
# This file is part of the assignment for the course    #
# "AI Driven Data Analytics Course" by Skillarbitrage   #
# Assignment ID: 748544                                 #
# Submissions of weekly Exercises for                   #
# Exercises on AI-Powered Automation for Data Science   # 
# Using GitHub Copilot in VS Code                       #
# Week Duration: 15/06/2025 to 22/06/2025               #
#                                                       #
# Student Name : Arindam Mitra                          #
# Registration Code: STU20250430/14045                  #
# Email: arinmitra@gmail.com                            #
#                                                       #
# Tutor - Mohit Sharma (mohit.sharma@skillarbitra.ge),  #
# Host - Ritu Verma (ritu.varma@skillarbitra.ge)        #
#########################################################

# Exercise 2: Creating an Automated Climate Data Dashboard with Copilot
# Task 4: Visualization – Building the Dashboard
""" 
Now create a series of visualizations that together form your climate “dashboard”. 
The idea is to convey the data over time and any notable patterns:

○ Temperature Over Time Line Chart:Plot the daily mean temperature over the 
entire period. This will result in a time-series line chart 
(dates on the x-axis, temperature on the y-axis) showing seasonal oscillations 
and year-to-year changes.

○ Ensure the x-axis has date ticks that are readable (for a multi-year range, 
you might see year labels or months). If the x-axis labels are cluttered, 
you can configure plt.xticks(rotation=45) or use matplotlib date formatting.

○ Observe the line chart: you should see a recurring yearly pattern (peaks in 
the middle of each year for summer, dips for winter). Also look for any overall 
trend – e.g., are the peaks getting higher each year or roughly the same?

"""

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# Load the dataset
file_path = 'cleaned_daily_delhi_climate.csv'
df = pd.read_csv(file_path)

# Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])
# Set the date as the index
df.set_index('date', inplace=True)

# Plotting the daily mean temperature over time
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['meantemp'], color='blue', label='Mean Temperature')
plt.title('Daily Mean Temperature in Delhi Over Time')
plt.xlabel('Date')
plt.ylabel('Mean Temperature (°C)')
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Save the plot as an image file
plt.savefig('daily_mean_temperature_delhi.png')

# Display the plot
plt.show()





```
